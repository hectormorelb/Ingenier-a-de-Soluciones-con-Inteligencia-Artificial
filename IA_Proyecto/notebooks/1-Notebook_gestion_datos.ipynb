# 📂 Paso 1: Cargar dataset

A continuación cargaremos el dataset original que contiene errores intencionales.

Antes de ejecutar, asegúrate de tener el archivo:

📄 ventas_raw.csv

en la misma carpeta del notebook.

```python
import shutil
import logging
import os

# Crear carpeta logs en la raíz del proyecto
os.makedirs("../logs", exist_ok=True)

logging.basicConfig(filename='../logs/ingest.log', level=logging.INFO)

# Rutas correctas desde notebooks/
origen = "ventas_raw.csv"
destino = "../data/raw/ventas_raw.csv"

try:
    shutil.copy(origen, destino)
    logging.info("Archivo copiado correctamente")
    print("Ingesta completada")
except Exception as e:
    logging.error(f"Error en ingesta: {e}")
```

```python
import pandas as pd

df = pd.read_csv("../data/raw/ventas_raw.csv")

df
```

# 🧹 Paso 3: Limpieza de datos

Aplicaremos las siguientes reglas:

- Eliminar duplicados
- Eliminar registros con valores nulos
- Corregir errores en precios
- Convertir tipos de datos

---

```python
# Eliminar duplicados
df = df.drop_duplicates()

# Eliminar nulos
df = df.dropna()

df
```

# 🔄 Paso 4: Corrección de tipos de datos

Convertiremos:

- precio → numérico
- fecha → formato datetime

```python
# Convertir precio a numérico
df["precio"] = pd.to_numeric(df["precio"], errors="coerce")

# Eliminar filas con valores nulos o errores de conversión
df = df.dropna().copy()

# Convertir fecha a tipo datetime
df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")

# Eliminar posibles errores generados en la conversión
df = df.dropna().copy()

df
```

# ⚙️ Paso 5: Transformación de datos

Crearemos una nueva columna:

📌 total = cantidad * precio

Esto permitirá analizar el valor de cada venta.

```python
df["total"] = df["cantidad"] * df["precio"]

df
```

# 💾 Paso 6: Guardar dataset limpio

El dataset final será almacenado para su uso posterior.

```python
df.to_csv("../data/processed/ventas_clean.csv", index=False)

print("Dataset limpio guardado correctamente")
```
